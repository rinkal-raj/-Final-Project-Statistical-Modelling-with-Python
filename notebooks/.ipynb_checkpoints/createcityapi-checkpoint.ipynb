**PART:1  Connecting to CityBikes API**

```python
import requests
import json
```

```python
#Define the end point URL for the milano city bike network
endpoint = "https://api.citybik.es/v2/networks/bikemi"
```

```python
# define the query parameter to include specific fields
params = {"fields":"location.latitude,location.longitude,stations.free_bikes"}
```

```python
#send https requests to url
response = requests.get(endpoint)
```

```python
#get json data  responce from request
json_data = response.json()
```

```python
#extract bike details from json format
bike_details = json_data["network"]["stations"]
```

```python
#print json data output
print(json.dumps(json_data, indent = 4))
```

```python
import pandas as pd
```

```python
milano_data = pd.DataFrame(bike_details, columns = ['name','latitude','longitude','free_bikes'])
```
