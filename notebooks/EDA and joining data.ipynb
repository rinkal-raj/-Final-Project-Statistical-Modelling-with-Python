```python
import requests
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline     
sns.set(color_codes=True)
from pathlib import Path
```

```python
merge_all_data = pd.read_csv('data/foursquar_park.csv')
merge_all_data.head(5)
```

```python
#checl all data type
merge_all_data.dtypes
```

```python
merge_all_data.shape
```

```python
duplicate_rows_df = merge_all_data[merge_all_data.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
merge_all_data.count()  
```

```python
merge_all_data = merge_all_data.drop_duplicates()
merge_all_data.head(5)
```

```python
merge_all_data.count()  
```

```python
print(merge_all_data.isnull().sum())
```

```python
merge_all_data = merge_all_data.dropna()    # Dropping the missing values.
merge_all_data.count()
```

```python
print(merge_all_data.isnull().sum())
```

```python
filepath = Path('EDA_data.csv')
filepath.parent.mkdir(parents=True, exist_ok=True)

merge_all_data.to_csv(filepath, index=False)
merge_all_data.head(5)
```

```python
#detecting outliners
sns.boxplot(x=merge_all_data['free_bikes'])
```

```python
sns.boxplot(x=merge_all_data['empty_slots'])
```

```python
sns.boxplot(x=merge_all_data['bike_station'])
```

```python
sns.boxplot(x=merge_all_data['rating'])
```

```python
#visualizaqtion for empty slot
sns.set_theme(context = 'notebook',     #recomend context = 'talk'
              style = 'darkgrid',       
              palette = 'deep',
              font = 'sans-serif',
              font_scale = 1
              ) 

# you could also set a figure size that will affect all sns plots in your notebook
#sns.set(rc={"figure.figsize":(10, 6)}) 

#seaborn histogram
sns.histplot(data=merge_all_data, x="empty_slots", bins=30)

plt.show()
```

```python
sns.kdeplot(data = merge_all_data, 
            x="rating", 
            fill=True)
```

```python
#relation between bikes and empty slot
sns.scatterplot(data = merge_all_data, 
                x = "free_bikes", 
                y = "empty_slots")
plt.suptitle("Scatter Plot Comparing empty slot to free bikes")

plt.show()
```

```python
# jointplot = scatter plot + 2 histograms
sns.jointplot(data = merge_all_data, 
              x = "free_bikes", 
              y = "rating", 
              height=8,
              ) # a jointplot will be square by default
plt.suptitle("JointPlot Comparing rating to review of resturant", y=1)

plt.show()
```
