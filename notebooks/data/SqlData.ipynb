```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect('data_park.db')
c = conn.cursor()
```

```python
# Run create table sql query
c.execute("create table if not exists park_details" +
             " (name text, location text, longitude integer,latitude integer)")
```

```python
# Load CSV data into Pandas DataFrame
park = pd.read_csv('foursquar_park.csv')
```

```python
# Write the data to a sqlite db table
park.to_sql('park_details', conn, if_exists='replace', index=False)
```

```python
# Run select sql query
c.execute('select * from park_details')
```

```python
records = c.fetchall()
```

```python
# Display result 
for row in records:
    # show row
    print(row)
```

```python
# Close connection to SQLite database
c.close()
```
